```python
# Importação das bibliotecas

import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

# Carga, agrupamento e união dos dados meteorológicos e de produção

```python
# Importação dos arquivos CSV dos dados meteorológicos e de produção

file = 'C:/Users/*******/Documents/Projetos Jupyter Notebook/TCC - Ciência de Dados/Dados_Meteorológicos.csv'
file2 = 'C:/Users/*******/Documents/Projetos Jupyter Notebook/TCC - Ciência de Dados/Dados_Produção.csv'
df_meteo = pd.read_csv(file, sep = ';')
df_prod = pd.read_csv(file2, sep = ';')
```

```python
 # Exibição dos dados meteorológicos
    
df_meteo.head()
```

```python
# Seleciona as colunas dos dados meteorológicos que serão utilizadas na análise

df_meteo = df_meteo.iloc[:, 2:17]
    
df_meteo.head() # Exibição dos dados
```

```python
# Agrupamento dos dados meteorológicos por safra

df_meteo_group = df_meteo.groupby('SAFRA').agg({'PRECIPITACAO TOTAL, DIARIO (AUT)(mm)' : 'sum',
                               'PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)' : 'mean',
                               'TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)' : 'mean',
                               'TEMPERATURA MAXIMA, DIARIA (AUT)(°C)' : 'mean',
                               'TEMPERATURA MEDIA, DIARIA (AUT)(°C)' : 'mean',
                               'TEMPERATURA MINIMA, DIARIA (AUT)(°C)' : 'mean',
                               'UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)' : 'mean',
                               'UMIDADE RELATIVA DO AR, MINIMA DIARIA (AUT)(%)' : 'mean',
                               'VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)' : 'mean',
                               'VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)' : 'mean',
                               'QTD. DE HORAS DE FRIO (< 7.2°C), DIARIA (MAIO - AGOSTO)':'sum',
                               'QTD. DE HORAS TEMP. IDEAL (18°C - 23 C°), DIARIA (SETEMBRO - ABRIL)':'sum',
                               'QTD. DE HORAS TEMP. PREJUDICIAL (> 27 C°), DIARIA (SETEMBRO - ABRIL)':'sum',
                               'PRECIPITACAO TOTAL, DIARIO (AUT)(mm) (SETEMBRO - ABRIL)':'sum'})

df_meteo_group # Exibição dos dados
```

```python
# Realização e visualização do ajuste das casas decimais dos dados

df_meteo_group['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'] = round(df_meteo_group['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'], 1)
df_meteo_group['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'] = round(df_meteo_group['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'], 1)
df_meteo_group['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'] = round(df_meteo_group['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'], 1)
df_meteo_group['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'] = round(df_meteo_group['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'], 1)
df_meteo_group['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'] = round(df_meteo_group['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'], 1)
df_meteo_group['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'] = round(df_meteo_group['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'], 1)
df_meteo_group['UMIDADE RELATIVA DO AR, MINIMA DIARIA (AUT)(%)'] = round(df_meteo_group['UMIDADE RELATIVA DO AR, MINIMA DIARIA (AUT)(%)'], 1)
df_meteo_group['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'] = round(df_meteo_group['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'], 1)
df_meteo_group['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'] = round(df_meteo_group['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'], 1)

df_meteo_group # Exibição dos dados
```

```python
# Remoção do índice dos dados

df_meteo_group = df_meteo_group.reset_index()

df_meteo_group # Exibição dos dados
```

```python
# Exibição dos dados de produção

df_prod.head()
```

```python
# Seleciona as colunas dos dados de produção que serão utilizadas na análise

df_prod = df_prod.iloc[:, 7:11]
df_prod.drop(["CALIBRE"], axis = 1, inplace = True)
    
df_prod.head() # Exibição dos dados
```

```python
# Agrupamento dos dados de produção por safra

df_prod_group = df_prod.groupby(['SAFRA', 'AGRUPAMENTO_TAMANHO']).agg({'QUANTIDADE_KG' : 'sum'})

df_prod_group # Exibição dos dados
```

```python
# Remoção do índice dos dados

df_prod_group = df_prod_group.reset_index()

df_prod_group # Exibição dos dados
```

```python
# União dos dados meteorológicos e de produção

df_merge = pd.merge(df_prod_group, df_meteo_group, how = 'outer', on = 'SAFRA').sort_values(by = 'SAFRA')

df_merge # Exibição dos dados
```

# Análise de correlação

```python
# Seleciona as colunas dos dados que serão utilizadas na análise

df_merge_agp_tam = df_merge.iloc[:, 1:]
    
df_merge_agp_tam.head() # Exibição dos dados
```

```python
# Extrai dos dados da lista de agrupamentos de tamanho da fruta

agp_tamanhos = df_merge_agp_tam.AGRUPAMENTO_TAMANHO.unique()
    
agp_tamanhos # Exibição dos dados
```

```python
# Executa a análise de correlação para cada grupo de tamanho do produto e armazena o resultado em um dataframe

cont = 0

for agp_tamanho in agp_tamanhos:
    df_agp_tamanho = df_merge_agp_tam.loc[df_merge_agp_tam['AGRUPAMENTO_TAMANHO'] == agp_tamanho]
    correlation = df_agp_tamanho.corr()
    plot = sn.heatmap(correlation, annot = True, fmt = ".1f", linewidths = .6)
    plt.title("Grupo de Tamanho: " + agp_tamanho)
    plt.show()
    if (cont == 0):
        df_corr_aux = correlation
        for index, row in df_corr_aux.iterrows():
            df_corr_aux.loc[index, 'GRUPO_TAMANHO'] = agp_tamanho
        df_corr = df_corr_aux    
    else:
        df_corr_aux = correlation
        for index, row in df_corr_aux.iterrows():
            df_corr_aux.loc[index, 'GRUPO_TAMANHO'] = agp_tamanho
        df_corr = pd.merge(df_corr, df_corr_aux, how = 'outer')
    cont = cont + 1
```

```python
df_corr # Exibição dos dados
```

```python
# Seleciona apenas os registros que relacionam a variável 'QUANTIDADE_KG' com as demais

df_corr_final = df_corr.loc[df_corr['QUANTIDADE_KG'] == 1]
df_corr_final # Exibição dos dados
```

```python
# Reset do índice dos dados

df_corr_final = df_corr_final.reset_index()
df_corr_final = df_corr_final.iloc[:, 1:]

df_corr_final # Exibição dos dados
```

```python
# Realização e visualização do ajuste das casas decimais dos dados

df_corr_final['QUANTIDADE_KG'] = round(df_corr_final['QUANTIDADE_KG'], 1)
df_corr_final['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'] = round(df_corr_final['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'], 1)
df_corr_final['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'] = round(df_corr_final['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'], 1)
df_corr_final['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'] = round(df_corr_final['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'], 1)
df_corr_final['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'] = round(df_corr_final['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'], 1)
df_corr_final['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'] = round(df_corr_final['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'], 1)
df_corr_final['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'] = round(df_corr_final['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'], 1)
df_corr_final['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'] = round(df_corr_final['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'], 1)
df_corr_final['UMIDADE RELATIVA DO AR, MINIMA DIARIA (AUT)(%)'] = round(df_corr_final['UMIDADE RELATIVA DO AR, MINIMA DIARIA (AUT)(%)'], 1)
df_corr_final['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'] = round(df_corr_final['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'], 1)
df_corr_final['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'] = round(df_corr_final['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'], 1)
df_corr_final['QTD. DE HORAS DE FRIO (< 7.2°C), DIARIA (MAIO - AGOSTO)'] = round(df_corr_final['QTD. DE HORAS DE FRIO (< 7.2°C), DIARIA (MAIO - AGOSTO)'], 1)
df_corr_final['QTD. DE HORAS TEMP. IDEAL (18°C - 23 C°), DIARIA (SETEMBRO - ABRIL)'] = round(df_corr_final['QTD. DE HORAS TEMP. IDEAL (18°C - 23 C°), DIARIA (SETEMBRO - ABRIL)'], 1)
df_corr_final['QTD. DE HORAS TEMP. PREJUDICIAL (> 27 C°), DIARIA (SETEMBRO - ABRIL)'] = round(df_corr_final['QTD. DE HORAS TEMP. PREJUDICIAL (> 27 C°), DIARIA (SETEMBRO - ABRIL)'], 1)
df_corr_final['PRECIPITACAO TOTAL, DIARIO (AUT)(mm) (SETEMBRO - ABRIL)'] = round(df_corr_final['PRECIPITACAO TOTAL, DIARIO (AUT)(mm) (SETEMBRO - ABRIL)'], 1)

df_corr_final
```
