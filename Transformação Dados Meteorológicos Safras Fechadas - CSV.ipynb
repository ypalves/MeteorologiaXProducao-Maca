```python
# Importação das bibliotecas

import pandas as pd
```

# Transformação dos dados meteorológicos diários

```python
# Importação do arquivo CSV dos dados meteorológicos diários

file = 'C:/Users/********/Documents/Projetos Jupyter Notebook/TCC - Ciência de Dados/Dados Meteorológicos Históricos - Vacaria/Por Dia/dados_A880_D_2009-05-01_2021-04-30.csv'
df = pd.read_csv(file, skiprows = 10, sep = ';')
df.head() # Exibição dos dados
```

```python
df.info() # Visualização do tipo de dados de cada coluna
```

```python
# Substituição de vírgulas por pontos nos dados das colunas com tipo não-numérico

df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'] = df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'].str.replace(',', '.')
df['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'] = df['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'].str.replace(',', '.')
df['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'] = df['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'].str.replace(',', '.')
df['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'] = df['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'].str.replace(',', '.')
df['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'] = df['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'].str.replace(',', '.')
df['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'] = df['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'].str.replace(',', '.')
df['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'] = df['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'].str.replace(',', '.')
df['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'] = df['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'].str.replace(',', '.')
df['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'] = df['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'].str.replace(',', '.')

df.head() # Exibição dos dados
```

```python
# Alteração do tipo de dados das colunas

df['Data Medicao'] = pd.to_datetime(df['Data Medicao'], dayfirst = True)
df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'] = pd.to_numeric(df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'])
df['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'] = pd.to_numeric(df['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'])
df['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'] = pd.to_numeric(df['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'])
df['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'] = pd.to_numeric(df['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'])
df['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'] = pd.to_numeric(df['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'])
df['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'] = pd.to_numeric(df['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'])
df['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'] = pd.to_numeric(df['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'])
df['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'] = pd.to_numeric(df['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'])
df['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'] = pd.to_numeric(df['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'])

df.info() # Visualização do tipo de dados de cada coluna
```

```python
# Realização e visualização do ajuste das casas decimais dos dados

df['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'] = round(df['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'], 1)
df['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'] = round(df['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'], 1)
df['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'] = round(df['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'], 1)
df['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'] = round(df['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'], 1)
df['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'] = round(df['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'], 1)

df.head()
```

```python
df.isnull().sum() # identificação de dados faltantes
```

```python
# Geração e visualização da média dos valores das colunas para utilizar na substituição dos valores nulos

precipitacao_media = df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'].mean()
pressao_atmosf_media = df['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'].mean()
temp_orvalho_media = df['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'].mean()
temp_maxima_media = df['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'].mean()
temp_media_media = df['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'].mean()
temp_minima_media = df['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'].mean()
umidade_media_media = df['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'].mean()
umidade_minima_media = df['UMIDADE RELATIVA DO AR, MINIMA DIARIA (AUT)(%)'].mean()
vento_maxima_media = df['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'].mean()
vento_media_media = df['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'].mean()

print('precipitacao_media: {}'.format(precipitacao_media))
print('pressao_atmosf_media: {}'.format(pressao_atmosf_media))
print('temp_orvalho_media: {}'.format(temp_orvalho_media))
print('temp_maxima_media: {}'.format(temp_maxima_media))
print('temp_media_media: {}'.format(temp_media_media))
print('temp_minima_media: {}'.format(temp_minima_media))
print('umidade_media_media: {}'.format(umidade_media_media))
print('umidade_minima_media: {}'.format(umidade_minima_media))
print('vento_maxima_media: {}'.format(vento_maxima_media))
print('vento_media_media: {}'.format(vento_media_media))
```

```python
# Realização e visualização do ajuste das casas decimais dos dados

precipitacao_media = round(precipitacao_media, 1)
pressao_atmosf_media = round(pressao_atmosf_media, 1)
temp_orvalho_media = round(temp_orvalho_media, 1)
temp_maxima_media = round(temp_maxima_media, 1)
temp_media_media = round(temp_media_media, 1)
temp_minima_media = round(temp_minima_media, 1)
umidade_media_media = round(umidade_media_media, 1)
umidade_minima_media = round(umidade_minima_media)
vento_maxima_media = round(vento_maxima_media, 1)
vento_media_media = round(vento_media_media, 1)

print('precipitacao_media: {}'.format(precipitacao_media))
print('pressao_atmosf_media: {}'.format(pressao_atmosf_media))
print('temp_orvalho_media: {}'.format(temp_orvalho_media))
print('temp_maxima_media: {}'.format(temp_maxima_media))
print('temp_media_media: {}'.format(temp_media_media))
print('temp_minima_media: {}'.format(temp_minima_media))
print('umidade_media_media: {}'.format(umidade_media_media))
print('umidade_minima_media: {}'.format(umidade_minima_media))
print('vento_maxima_media: {}'.format(vento_maxima_media))
print('vento_media_media: {}'.format(vento_media_media))
```

```python
# Substituição dos dados nulos e verificação se realmente não restou nenhum dado nulo

df.update(df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'].fillna(precipitacao_media))
df.update(df['PRESSAO ATMOSFERICA MEDIA DIARIA (AUT)(mB)'].fillna(pressao_atmosf_media))
df.update(df['TEMPERATURA DO PONTO DE ORVALHO MEDIA DIARIA (AUT)(°C)'].fillna(temp_orvalho_media))
df.update(df['TEMPERATURA MAXIMA, DIARIA (AUT)(°C)'].fillna(temp_maxima_media))
df.update(df['TEMPERATURA MEDIA, DIARIA (AUT)(°C)'].fillna(temp_media_media))
df.update(df['TEMPERATURA MINIMA, DIARIA (AUT)(°C)'].fillna(temp_minima_media))
df.update(df['UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)'].fillna(umidade_media_media))
df.update(df['UMIDADE RELATIVA DO AR, MINIMA DIARIA (AUT)(%)'].fillna(umidade_minima_media))
df.update(df['VENTO, RAJADA MAXIMA DIARIA (AUT)(m/s)'].fillna(vento_maxima_media))
df.update(df['VENTO, VELOCIDADE MEDIA DIARIA (AUT)(m/s)'].fillna(vento_media_media))

df.isnull().sum()
```

# Transformação dos dados meteorológicos horários

```python
# Importação do arquivo CSV dos dados meteorológicos horários

file = 'C:/Users/*********/Documents/Projetos Jupyter Notebook/TCC - Ciência de Dados/Dados Meteorológicos Históricos - Vacaria/Por Hora/dados_A880_H_2009-05-01_2021-04-30.csv'
df_hora = pd.read_csv(file, skiprows = 10, sep = ';')
df_hora.head() # Exibição dos dados
```

```python
df_hora = df_hora.iloc[:, [0, 9]] # Seleciona as colunas 'Data Medicao' e 'TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'
df_hora.head() # Exibição dos dados
```

```python
df_hora.info() # Visualização do tipo de dados de cada coluna
```

```python
# Substituição de vírgulas por pontos nos dados das colunas com tipo não-numérico

df_hora['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'] = df_hora['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'].str.replace(',', '.')

df_hora.head() # Exibição dos dados
```

```python
# Alteração do tipo de dados das colunas

df_hora['Data Medicao'] = pd.to_datetime(df_hora['Data Medicao'], dayfirst = True)
df_hora['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'] = pd.to_numeric(df_hora['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'])

df_hora.info() # Visualização do tipo de dados de cada coluna
```

```python
df_hora.isnull().sum() # identificação de dados faltantes
```

```python
# Geração e visualização da média dos valores da coluna que possui dados faltantes para utilizar na substituição dos 
# valores nulos

temp_bulbo_media = df_hora['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'].mean()

print('temp_bulbo_media: {}'.format(temp_bulbo_media))
```

```python
# Realização e visualização do ajuste das casas decimais dos dados

temp_bulbo_media = round(temp_bulbo_media, 1)

print('temp_bulbo_media: {}'.format(temp_bulbo_media))
```

```python
# Substituição dos dados nulos e verificação se realmente não restou nenhum dado nulo

df_hora.update(df_hora['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'].fillna(temp_bulbo_media))

df_hora.isnull().sum()
```

```python
# Seleção dos registros com temperatura abaixo de 7.2°C (horas de frio) no ciclo de dormência da planta (Maio a agosto)

df_hora_frio = df_hora.loc[df_hora['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'] < 7.2]

df_hora_frio = df_hora_frio.loc[df_hora_frio['Data Medicao'].dt.month.between(5, 8)]

df_hora_frio # Exibição dos dados
```

```python
# Contagem/agrupamento das horas de frio diárias e alteração do nome da coluna

df_hora_frio = df_hora_frio.groupby(['Data Medicao']).count()

df_hora_frio.rename(columns = {'TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)': 'QTD. DE HORAS DE FRIO (< 7.2°C), DIARIA (MAIO - AGOSTO)'}, inplace = True)

df_hora_frio.head()
```

```python
# União do dataset com o número diário de horas de frio com o dataset dos dados meteorológicos diários

df = pd.merge(df, df_hora_frio, how = 'outer', on = 'Data Medicao').sort_values(by = 'Data Medicao')
```

```python
# Substituição dos dados nulos pelo número 0 na coluna das horas de frio

df.update(df['QTD. DE HORAS DE FRIO (< 7.2°C), DIARIA (MAIO - AGOSTO)'].fillna(0))
```

```python
# Alteração do tipo de dados da coluna das horas de frio para número inteiro e exibição da alteração

df['QTD. DE HORAS DE FRIO (< 7.2°C), DIARIA (MAIO - AGOSTO)'] = pd.to_numeric(df['QTD. DE HORAS DE FRIO (< 7.2°C), DIARIA (MAIO - AGOSTO)'], downcast = 'integer')
df.head()
```

```python
# Seleção dos registros de temperatura do ciclo vegetativo da planta (Setembro a abril)

df_hora_tempcv = df_hora.loc[df_hora['Data Medicao'].dt.month.isin([1, 2, 3, 4, 9, 10, 11, 12])]

df_hora_tempcv # Exibição dos dados
```

```python
# Contagem/agrupamento das horas de temperatura ideal para o ciclo vegetativo (18°C a 23 C°) e alteração do nome da coluna

df_hora_idealcv = df_hora_tempcv[df_hora_tempcv['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'].between(18, 23)]

df_hora_idealcv = df_hora_idealcv.groupby(['Data Medicao']).count()

df_hora_idealcv.rename(columns = {'TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)': 'QTD. DE HORAS TEMP. IDEAL (18°C - 23 C°), DIARIA (SETEMBRO - ABRIL)'}, inplace = True)

df_hora_idealcv.head()
```

```python
# União do dataset com o número diário de horas de temperatura ideal para o ciclo vegetativo com o dataset dos dados meteorológicos diários

df = pd.merge(df, df_hora_idealcv, how = 'outer', on = 'Data Medicao').sort_values(by = 'Data Medicao')
```

```python
# Substituição dos dados nulos pelo número 0 na coluna das horas de temperatura ideal para o ciclo vegetativo

df.update(df['QTD. DE HORAS TEMP. IDEAL (18°C - 23 C°), DIARIA (SETEMBRO - ABRIL)'].fillna(0))
```

```python
# Alteração do tipo de dados da coluna das horas de temperatura ideal para o ciclo vegetativo para número inteiro e exibição da alteração

df['QTD. DE HORAS TEMP. IDEAL (18°C - 23 C°), DIARIA (SETEMBRO - ABRIL)'] = pd.to_numeric(df['QTD. DE HORAS TEMP. IDEAL (18°C - 23 C°), DIARIA (SETEMBRO - ABRIL)'], downcast = 'integer')
df.head()
```

```python
# Contagem/agrupamento das horas de temperatura prejudicial para o ciclo vegetativo (> 27 C°) e alteração do nome da coluna

df_hora_prejucv = df_hora_tempcv[df_hora_tempcv['TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)'] > 27]

df_hora_prejucv = df_hora_prejucv.groupby(['Data Medicao']).count()

df_hora_prejucv.rename(columns = {'TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)': 'QTD. DE HORAS TEMP. PREJUDICIAL (> 27 C°), DIARIA (SETEMBRO - ABRIL)'}, inplace = True)

df_hora_prejucv.head()
```

```python
# União do dataset com o número diário de horas de temperatura prejudicial para o ciclo vegetativo com o dataset dos dados meteorológicos diários

df = pd.merge(df, df_hora_prejucv, how = 'outer', on = 'Data Medicao').sort_values(by = 'Data Medicao')
```

```python
# Substituição dos dados nulos pelo número 0 na coluna das horas de temperatura prejudicial para o ciclo vegetativo

df.update(df['QTD. DE HORAS TEMP. PREJUDICIAL (> 27 C°), DIARIA (SETEMBRO - ABRIL)'].fillna(0))
```

```python
# Alteração do tipo de dados da coluna das horas de temperatura prejudicial para o ciclo vegetativo para número inteiro e exibição da alteração

df['QTD. DE HORAS TEMP. PREJUDICIAL (> 27 C°), DIARIA (SETEMBRO - ABRIL)'] = pd.to_numeric(df['QTD. DE HORAS TEMP. PREJUDICIAL (> 27 C°), DIARIA (SETEMBRO - ABRIL)'], downcast = 'integer')
df.head()
```

```python
# Seleção dos registros de precipitação do ciclo vegetativo da planta (Setembro a abril)

df_prec_veg = df.loc[df['Data Medicao'].dt.month.isin([1, 2, 3, 4, 9, 10, 11, 12])]

df_prec_veg # Exibição dos dados
```

```python
df_prec_veg = df_prec_veg.iloc[:, [0, 1]] # Seleciona as colunas 'Data Medicao' e 'PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'
df_prec_veg.head() # Exibição dos dados
```

```python
df_prec_veg.rename(columns = {'PRECIPITACAO TOTAL, DIARIO (AUT)(mm)': 'PRECIPITACAO TOTAL, DIARIO (AUT)(mm) (SETEMBRO - ABRIL)'}, inplace = True)

df_prec_veg.head()
```

```python
# União do dataset com a precipitação do ciclo vegetativo da planta com o dataset dos dados meteorológicos diários

df = pd.merge(df, df_prec_veg, how = 'outer', on = 'Data Medicao').sort_values(by = 'Data Medicao')
```

```python
# Substituição dos dados nulos pelo número 0 na coluna da precipitação do ciclo vegetativo da planta

df.update(df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm) (SETEMBRO - ABRIL)'].fillna(0))
```

```python
# Alteração do tipo de dados da coluna da precipitação do ciclo vegetativo da planta para número inteiro e exibição da alteração

df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm) (SETEMBRO - ABRIL)'] = pd.to_numeric(df['PRECIPITACAO TOTAL, DIARIO (AUT)(mm) (SETEMBRO - ABRIL)'], downcast = 'integer')
df.head()
```

```python
# Geração de uma nova coluna que indica a que safra o registro pertence (É considerado como ano agrícola o período entre 
# 01 de maio até 30 de abril do próximo ano)

for index, row in df.iterrows():
    if df.loc[index, 'Data Medicao'].month in (1, 2, 3, 4):
        df.loc[index, 'SAFRA'] = str(df.loc[index, 'Data Medicao'].year)
    else:
        df.loc[index, 'SAFRA'] = str(df.loc[index, 'Data Medicao'].year + 1)
    
df.head() # Exibição dos dados
```
